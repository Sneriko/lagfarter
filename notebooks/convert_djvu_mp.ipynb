```python
import multiprocessing as mp
from multiprocessing import Pool
import subprocess
from glob import glob
import os
from pathlib import Path
```

```python
def run_script(arg_str):
    command = '/home/erik/Riksarkivet/Projects/lagfarter/bash/./convert_djvu_mp.sh'
    files = glob(os.path.join(arg_str, '**'))
    files.sort()
    for file in files:
        #print("Starting command :{} with argument {}".format(command, ))
        subprocess.call(command+" "+file, shell=True)
    #print("Completed batch: {}".format(arg_str))
```

```python
def make_dirs():
    out_path = '/home/erik/Riksarkivet/Projects/lagfarter/data/indexed_batches_zip2_tiff'
    dirs = glob(os.path.join('/home/erik/Riksarkivet/Projects/lagfarter/data/indexed_batches_djvu', '**'))
    for dir in dirs:
        out_dir = os.path.join(out_path, Path(dir).name)
        os.makedirs(out_dir, exist_ok=True)
```

```python
make_dirs()

batches = glob(os.path.join('/home/erik/Riksarkivet/Projects/lagfarter/data/indexed_batches_djvu', '**'))
batches.sort()

with Pool(mp.cpu_count()) as p:
    p.map(run_script, batches)

```
