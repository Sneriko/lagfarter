```python
import json
import mmcv
import os
```

```python
with open('/home/erik/Riksarkivet/Projects/lagfarter/data/det_it2_ls_export_2500/result_7000.json', 'r') as f:
    coco_2500 = json.load(f)

with open('/home/erik/Riksarkivet/Projects/lagfarter/data/lagfart_det_7000.json', 'r') as f:
    exp_7000 = json.load(f)
```

```python
coco = dict()
coco['images'] = list()
coco['categories'] = [
    {
      "id": 0,
      "name": "Pagenr"
    },
    {
      "id": 1,
      "name": "Property"
    },
    {
      "id": 2,
      "name": "Propnr"
    },
    {
      "id": 3,
      "name": "Socken"
    }
  ]
coco['annotations'] = list()

path_to_imgs = '/home/erik/Riksarkivet/Projects/lagfarter/data/det_it2_ls_export_2500/images'


category_dict = {
    'Pagenr': 0,
    'Property': 1,
    'Propnr': 2,
    'Socken': 3 
    }

for i, image in enumerate(exp_7000):
    file_name = image['data']['image'].split('/')[-1].split('-')[1]
    img_p = os.path.join(path_to_imgs, file_name)
    try:
        img = mmcv.imread(img_p)
    except:
        continue
    img_dict = dict()
    width = img.shape[1]
    height = img.shape[0]

    img_dict['width'] = width
    img_dict['height'] = height
    img_dict['id'] = i
    img_dict['file_name'] = file_name

    coco['images'].append(img_dict)

    for annot in image['annotations']:
        for ann in annot['result']:
            if ann['type'] == 'rectanglelabels':
                pixel_x = ann['value']['x'] / 100 * width
                pixel_y = ann['value']['y'] / 100 * height
                pixel_width = ann['value']['width'] / 100 * width
                pixel_height = ann['value']['height'] / 100 * height
                image_id = i
                area = pixel_width * pixel_height
                category_id = category_dict[ann['value']['rectanglelabels'][0]]

                ann_dict = dict()
                ann_dict['image_id'] = image_id
                ann_dict['category_id'] = category_id
                ann_dict['bbox'] = [pixel_x, pixel_y, pixel_width, pixel_height]
                ann_dict['ignore'] = 0
                ann_dict['iscrowd'] = 0
                ann_dict['area'] = area

                coco['annotations'].append(ann_dict)

    if i % 100 == 0:
        print(i)
    
```

```python
for j, annot in enumerate(coco['annotations']):
    coco['annotations'][j]['id'] = j
```

```python
#remove "28" from coco_2500

for i, img in enumerate(coco_2500['images']):
    new_filename = img['file_name'].split('/')
    new_filename.remove('28')
    new_filename = '/'.join(new_filename)
    coco_2500['images'][i]['file_name'] = new_filename
```

```python
with open('/home/erik/Riksarkivet/Projects/lagfarter/data/det_it2_ls_export_2500/coco_7000.json', 'w') as f:
    json.dump(coco, f, indent=2, ensure_ascii=False)
```
